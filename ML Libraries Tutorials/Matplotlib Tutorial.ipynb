## Matplotlib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics
extension NumPy. It provides an object-oriented API for embedding plots into applications using
general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+

Some of the major PROs of Matplotlib are:

1. Generally easy to get started for simpler plots
2. Support for custom labels and texts
3. Great control of every element in a figure
4. High quality output in many formats
5. Very customizable in general

```python
import matplotlib.pyplot as plt

# %matplotlib inline is written in order to avoid writting plt.show()
%matplotlib inline
```

```python
import numpy as np
```

```python
# Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
# Plotting using matplotlib

# Scatter Plot

plt.scatter(x,y,c='r')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2-D Graph')
plt.savefig('test.png')
```

```python
y=x*x
```

```python
# Plot

plt.plot(x,y,'g*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2-D Graph')
plt.savefig('test1.png')
```

```python
# Creating Subplots

# plt.subplot(rows,columns,position)

plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'go')

plt.subplot(2,2,3)
plt.plot(x,y,'b*')

plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x,y)
```

```python
# Compute the x and y co-ordinates for points on a sine curve

x=np.arange(0,4 * np.pi, 0.1)
y=np.sin(x)
plt.title("Sine Wave")
plt.plot(x,y)
```

```python
# Subplot

# Compute the x and y co-ordinates for points on sine and cosine curves

x=np.arange(0,5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid such that it has height 2 and width 1
# And set the first such subplot as active
plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')

# Set the second subplot as active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

# Show the figure
plt.show()
```

```python
# Bar Plot

x1 = [2,8,10]
y1 = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x1,y1,color='g')
plt.bar(x2,y2,color='r')
plt.title('Bar Graph')
plt.xlabel('X-Axix')
plt.ylabel('Y-Axix')

plt.show()
```

## Histogram

```python
a = np.array(np.random.randint(0,50,20))
plt.hist(a,bins=20)
plt.title('Histogram')
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

# Rectangular Box Plot
plt.boxplot(data,vert=True,patch_artist=True);
```

## Pie Chart

```python
# Data to plot

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1,0,0,0) # Explode first slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
